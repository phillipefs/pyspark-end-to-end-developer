```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .getOrCreate()

sc = spark.sparkContext
```

Shuffle<br>
- Shuffling is a process of redistributing data across partitions or even nodes.
- Shuffle operation would create a new stage. 
- Based on data size we may reduce or increase the number os partitions using configuration spark.sql.shuffle.partitions or through codes like 
repartition and coalesce.
- Costly operation as it involves disk I/O, network I/O and data serialization /de-serialization.
- Spark shuffling tiggers for transformations operations lie gropByKey(), reduceByKey(), join(), union(), cogroup(), groupBy() etc ..
- Distinct creates a shuffle. 
- Count and countByKey does not create any shuffle.

Combiner<br>
- It computes the intermediate values for each partition to avoid partial shuffling.
